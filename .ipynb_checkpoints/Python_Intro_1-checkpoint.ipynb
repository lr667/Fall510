## Characters

```python
s = 'spam'                # Assign the string 'spam' to the variable s.
print(s[0], s[-2])        # Print the first and second last character of s
print(s[1:2], s[1:], s[:-1]) # Print the second to third character of s; the second to the last character of s; the first to the second last of s
```

### Import variables from a module

```python
import Class_1_Module # use Class_1_Module name space
dir(Class_1_Module)  # what is the meaning of dir(), and which situation shall I use it?
# SW: https://docs.python.org/3/library/functions.html#dir
print(Class_1_Module.title) # print the title of Class_1_Module
print(Class_1_Module.MyAge) # print the calcilated answer of Class_1_Module, that is, MyAge.
# Module is like a class in Java, include some functions, return some variables.
help(dir)  # Use help() to get the meaning of dir().
```

## Show the difference in scope between importing a module and importing a specific variable from the module


```python
#copies into local name space, overwrites same named variable
#without warning
from Class_1_Module import title 
print(title)
# The title is variable, and we just import it from the Class_1_Module name space, our original definition of title 
# is overwritten. In this way, if we want, we can assign a new name to it such as classnote or others.
from Class_1_Module import title as classnote
# for example, we can try to change the title like this.
```
